```python
import tkinter as tk

# Initialize the main window
window = tk.Tk()
window.title("Buttons with Radio Buttons")
window.geometry("700x400")

# Create Buttons (A to F) with different background colors
buttons = []
button_labels = ["Load PIB/PIF data", "Plot raw data", "Plot normalized data", "Save normalized data", "Save PIB/PIF segment", "Exit Viewer"]
button_colors = ["white", "blue", "green", "yellow", "orange", "purple"]

for i in range(6):
    button = tk.Button(window, text=button_labels[i], width=16, bg=button_colors[i])
    button.grid(row=0, column=i, padx=5, pady=20)
    buttons.append(button)

# Variables for Radio Buttons
radio_var_a = tk.StringVar(value="OptionA")
radio_var_b = tk.StringVar(value="OptionB")

# Create and Place Radio Buttons for Button A
radio1_a = tk.Radiobutton(window, variable=radio_var_a, value="a1")
radio2_a = tk.Radiobutton(window, variable=radio_var_a, value="a2")
radio3_a = tk.Radiobutton(window, variable=radio_var_a, value="a3")

radio1_a.grid(row=1, column=0, padx=5, sticky="w")
radio2_a.grid(row=1, column=0)
radio3_a.grid(row=1, column=0, padx=5, sticky="e")

# Add Labels under Radio Buttons for A
label1_a = tk.Label(window, text="a1")
label2_a = tk.Label(window, text="a2")
label3_a = tk.Label(window, text="a3")

label1_a.grid(row=2, column=0, padx=5, sticky="w")
label2_a.grid(row=2, column=0)
label3_a.grid(row=2, column=0, padx=5, sticky="e")

# Create and Place Radio Buttons for Button B
radio1_b = tk.Radiobutton(window, variable=radio_var_b, value="b1")
radio2_b = tk.Radiobutton(window, variable=radio_var_b, value="b2")
radio3_b = tk.Radiobutton(window, variable=radio_var_b, value="b3")

radio1_b.grid(row=1, column=1, padx=5, sticky="w")
radio2_b.grid(row=1, column=1)
radio3_b.grid(row=1, column=1, padx=5, sticky="e")

# Add Labels under Radio Buttons for B
label1_b = tk.Label(window, text="b1")
label2_b = tk.Label(window, text="b2")
label3_b = tk.Label(window, text="b3")

label1_b.grid(row=2, column=1, padx=5, sticky="w")
label2_b.grid(row=2, column=1)
label3_b.grid(row=2, column=1, padx=5, sticky="e")

# Run the main loop
window.mainloop()
```
